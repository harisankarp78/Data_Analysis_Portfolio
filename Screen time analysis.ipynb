# I am going to analyse screen time using a data set I got from Kaggle.
#### For this I have used matplotlib and seaborn as a visualization library.Seaborn itself cannot handle the plot size so I combined matplotlib and seaborn together

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import seaborn as sns
print("completed")
```

```python
df= pd.read_csv('Screentime-App-Details.csv')
```

```python
df.head()
```

## So the data set contains date, usage, Notifications, Times opened, App details.

```python
df.isnull().sum()
```

## We found that there is no null value in the data set

```python
df.describe()
```

## Now I am going to visualize certain parts

```python
df.App.unique()
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(30, 10))
sns.barplot(data=df, x= "Date", y="Usage", hue="App",width=0.8,ax=ax).set(title="Usage per date per app")
plt.show()
```

```python
fig, ax= plt.subplots(figsize=(30,10))
sns.barplot (df,x="Date",y="Notifications",hue="App",ax=ax).set(title='Notification per day per App')
```

```python
fig, ax=plt.subplots(figsize=(30,10))
sns.barplot(df,x="Date",y="Times opened",hue="App",ax=ax).set(title="Opened")
```

## I want to check whether there is any relationship between number of notifications received and phone usage. So I just plotted a reggression plot using seaborn below

```python
sns.regplot (df,x="Notifications",y="Times opened")
```

## The plot shows that there is a linear relationship between number of notification and phone usage
